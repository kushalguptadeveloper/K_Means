```python
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
```

```python
x,y = make_blobs()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x[:,0],x[:,1],c = y)
```

```python
mergings = linkage(x, method='complete')
```

```python
dendrogram(mergings,
           labels=y
           )
plt.show()
```
