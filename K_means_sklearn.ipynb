```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[1,2], [1.5,1.8], [5,8], [8,8], [1,0.6], [9,11]])
```

```python
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
k_means = KMeans(n_clusters = 2)
```

```python
k_means.fit(X)
```

```python
k_means.labels_
```

```python
k_means.cluster_centers_
```

```python
plt.scatter(X[:,0], X[:,1], c=k_means.labels_)
plt.scatter(k_means.cluster_centers_[:,0],k_means.cluster_centers_[:,1])
plt.show()
```

# Applying k-means on iris

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
k_means_iris = KMeans(n_clusters = 3)
```

```python
k_means_iris.fit(x)
```

```python
#k_means_iris.score(k_means_iris.labels_,y)
#l = k_means_iris.labels_.reshape(-1,1)
plt.scatter(x[:,0],x[:,1],c = k_means_iris.labels_)
```

```python
from sklearn.metrics import accuracy_score
```

```python
#accuracy_score(y,k_means_iris.labels_)
yy = []
pred = k_means_iris.labels_
for i in range(len(pred)):
    if pred[i] == 1:
        yy.append(0)
    elif pred[i] == 2:
        yy.append(1)
    elif pred[i] == 0:
        yy.append(2)
```

```python
yy = np.array(yy)
```

```python
accuracy_score(y,yy)
```
